From the graphs we see that there are obvious up and downs with respect to time, so we can impute the null values in Daily Mean Travel Time by incrementing or decrementing nearby values.

```python
#graph the travel from Powell to The palace
df = bart_origins
pow_pal = df.loc[(df['Origin Movement ID'] == 3760) & 
                 (df['Destination Movement ID'] == 3396)][['Date', 'Daily Mean Travel Time (Seconds)']]
pow_pal = pow_pal.sort_values('Date').reset_index(drop=True)

x = pow_pal.loc[range(181),'Date']
y = pow_pal.loc[range(181),'Daily Mean Travel Time (Seconds)']

# plot
plt.plot(x,y)
# beautify the x-labels
plt.gcf().autofmt_xdate()

plt.show()
```

```python
id_name_dict
```

```python
#Imputation
new_pow_pal = pow_pal.set_index('Date').interpolate('time').reset_index()

x = new_pow_pal.loc[range(180),'Date']
#x = range(181)
y = new_pow_pal.loc[range(180),'Daily Mean Travel Time (Seconds)'].diff()
# plot
plt.plot(x,y)
# beautify the x-labels
plt.gcf().autofmt_xdate()

plt.show()
```

```python
new_pow_pal
```

```python
impute = lambda x: new_pow_pal.loc[new_pow_pal['Date'] == x['Date']]['Daily Mean Travel Time (Seconds)'].iloc[0]
for i in range(len(bart_origins)):
    row = bart_origins.iloc[i]
    if (row['Origin Movement ID'] == 3760 and
        row['Destination Movement ID'] == 3396 and
        np.isnan(row['Daily Mean Travel Time (Seconds)'])):
        imputed = impute(row)
        print(imputed)
        print(bart_origins.iloc[i]['Daily Mean Travel Time (Seconds)'])
        bart_origins.set_value(i, 'Daily Mean Travel Time (Seconds)', imputed)
        print('imputed:' + str(bart_origins.iloc[i]['Daily Mean Travel Time (Seconds)']))
bart_origins
```

# Building Models

We will build a model that takes in the date attribute and outputs the travelling time.

```python
import warnings
import itertools
from pylab import rcParams
import statsmodels.api as sm
```

```python
y = bart_origins[['Date', 'Daily Mean Travel Time (Seconds)']].set_index('Date')
y
```

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(0, 0, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(18, 8))
plt.show()
```
